#### Для начала преобразуем список слов в словарь

```python
def file_to_dict():
    filename = 'wg.txt'
    with open(filename) as f:
            lines = f.readlines()
    lines2 = []
    for i in lines:
        lines2.append(i[:-1])
    d = {}
    for i in lines2:
        key = i[0]
        if key not in d:
            d[key] = [i]
        else: d[key].append(i)
    return d
```
